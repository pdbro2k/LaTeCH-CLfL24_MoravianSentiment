```python
import pandas as pd
```

# Load Data

```python
data_path = "../../data/fine-tuning/"
```

```python
train_df = pd.read_csv(data_path + "moravian_deu_train.csv")
train_df
```

```python
test_df = pd.read_csv(data_path + "moravian_deu_test.csv") 
test_df
```

```python
list(test_df["label"].value_counts())
```

# Compile Overview

```python
data = {
    "Train": reversed(train_df["label"].value_counts()),
    "Test": reversed(test_df["label"].value_counts()),
}

overview_df = pd.DataFrame.from_dict(data, orient="index", columns=["Negative_Instances", "Neutral_Instances", "Positive_Instances"])
overview_df["Total"] = overview_df.sum(axis=1)
overview_df.loc['Total',:] = overview_df.sum(axis=0)
overview_df
```
