# Compile Overview

```python
def get_scores(y_true, y_predicted):
    return [accuracy_score(y_true, y_predicted)] + list(f1_score(y_true, y_predicted, average=None))
```

```python
data = {
    "Majority Baseline": get_scores(y_true, y_majority),
    "SentiWS": get_scores(y_true, y_sentiws),
    "GerVADER": get_scores(y_true, y_gervader),
    "ger-senti-bert": get_scores(y_true, y_ger_senti_bert),
    "senti-distilbert": get_scores(y_true, y_senti_distilbert),
    "bert*": get_scores(y_true, y_finetuned_bert),
    "gbert*": get_scores(y_true, y_finetuned_gbert),
    "ger-senti-bert*": get_scores(y_true, y_finetuned_ger_senti_bert)
}

pd.DataFrame.from_dict(data, orient="index", columns=["Accuracy", "F1_negative", "F1_neutral", "F1_positive"])
```
