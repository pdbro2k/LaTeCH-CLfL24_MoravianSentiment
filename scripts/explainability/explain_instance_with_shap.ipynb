```python
import pandas as pd

import shap

import torch
from transformers import pipeline, BertTokenizerFast
```

# Init Instance

```python
instance = ["Die lezte Zeit kr√§nckelte er."]
```

# Init Transformers

```python
gpu = -1
if torch.cuda.is_available():
    torch.device("cuda")
    gpu = torch.cuda.current_device()
```

# Analyze GBERT

```python
gbert_model_path = "../../models/fine-tuned/gbert-base-moravian-sentiment"
gbert_model = pipeline("text-classification", model=gbert_model_path, device=gpu, top_k=None)
gbert_tokenizer = BertTokenizerFast.from_pretrained(gbert_model_path)

gbert_explainer = shap.Explainer(gbert_model, gbert_tokenizer)
```

```python
gbert_shap_values = gbert_explainer(instance)

# visualize the explanations for all output classes
shap.plots.text(gbert_shap_values)
```

# Analyze BERT

```python
bert_model_path = "../../models/fine-tuned/bert-base-cased-moravian-sentiment"
bert_model = pipeline("text-classification", model=bert_model_path, device=gpu, top_k=None)
bert_tokenizer = BertTokenizerFast.from_pretrained(bert_model_path)

bert_explainer = shap.Explainer(bert_model, bert_tokenizer)
```

```python
bert_shap_values = bert_explainer(instance)

# visualize the explanations for all output classes
shap.plots.text(bert_shap_values)
```
