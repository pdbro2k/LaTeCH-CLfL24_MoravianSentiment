```python
from geodatasets import get_path
import geopandas

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

# Load Data

```python
data_path = "../../data/fine-tuning/"
```

```python
metadata_df = pd.read_csv(data_path + "moravian_deu_metadata.csv", index_col=0)
metadata_df
```

# Add Geometry

```python
metadata_df = geopandas.GeoDataFrame(
    metadata_df, geometry=geopandas.points_from_xy(metadata_df["birthplace_long"], metadata_df["birthplace_lat"]), crs="EPSG:4326"
)
metadata_df
```

# Draw Map

```python
# get background map
world = geopandas.read_file(get_path("naturalearth.land"))

ax = world.clip([-100, 0, 40, 70]).plot(color="lightgrey", edgecolor="grey", figsize=(20,10))
metadata_df.plot(ax=ax, color="blue")
plt.tick_params(left=False, bottom=False, labelleft=False, labelbottom=False)
plt.show()
```
